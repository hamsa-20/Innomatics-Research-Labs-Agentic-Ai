```python

#  User Login Check


print("----- 1. User Login Check -----")

username = "admin"
password = "1234"

input_username = "admin"      # You can change these values for testing
input_password = "1234"

if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")

print("\n")


#  Pass / Fail Analyzer


print("----- 2. Pass / Fail Analyzer -----")

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Total Passed:", pass_count)
print("Total Failed:", fail_count)

print("\n")



# Simple Data Cleaner


print("----- 3. Simple Data Cleaner -----")

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Original Names:", names)
print("Cleaned Names:", cleaned_names)

print("\n")

# Message Length Analyzer


print("----- 4. Message Length Analyzer -----")

messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")
    
    if length > 10:
        print(" This message is longer than 10 characters.")

print("\n")



# Error Message Detector


print("----- 5. Error Message Detector -----")

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")

print("Total ERROR messages:", error_count)

print("\n")
print("----- Task Completed Successfully -----")
```
