```python
# Problem 1: Employee Performance Bonus Eligibility

# Dictionary containing employee names and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest performance score
highest_score = max(employees.values())

# Step 2: Find all employees who have the highest score (handle ties)
top_performers = [name for name, score in employees.items() if score == highest_score]

# Step 3: Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
# Problem 2: Search Query Keyword Analysis

# Input search query
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation (if any)
import string
query = query.translate(str.maketrans("", "", string.punctuation))

# Step 3: Split into words
words = query.split()

# Step 4: Count frequency of each keyword
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Step 5: Filter words appearing more than once
frequent_words = {word: count for word, count in word_count.items() if count > 1}

# Display result
print(frequent_words)
```

```python
# Problem 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Step 1: Identify even readings
valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:  # Check if even
        valid_readings.append((index, value))

# Step 2: Display valid readings
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Extract domains
domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Step 2: Calculate percentage usage
total_emails = len(emails)

print("Email Domain Usage Percentage:")
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average sales
average_sales = sum(sales) / len(sales)

# Step 2: Detect sales more than 30% above average
threshold = average_sales * 1.3

print("Sales Spikes (More than 30% above average):")

for day, sale in enumerate(sales, start=1):
    if sale > threshold:
        print(f"Day {day}: {sale}")
```

```python
# Problem 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Count occurrences
id_count = {}

for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

# Step 2: Display duplicates only
print("Duplicate User IDs:")

for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
